# **Zomato Dataset Exploratory Data Analysis**

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=True,cmap='viridis')
```

```python
df_country=pd.read_excel('/kaggle/input/dataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.head
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y='Rating Count',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```
